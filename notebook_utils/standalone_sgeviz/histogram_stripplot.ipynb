# Histogram + Strip Plot

Plots the distribution of SGE fitness scores as a histogram (top) and a strip plot of individual variants (bottom), with threshold lines overlaid.

**Required data:** `scores` and `thresholds` sheets from the pipeline output Excel file (`*_data.xlsx`).

**Required packages:** `sge-utils` (SimpleSGEViz) installed in the active environment.

**Saving:** PNG/SVG require `vl-convert-python` (`pip install vl-convert-python`). Change the extension to `.html` for interactive output (no extra package needed).

```python
import pandas as pd
from sgeviz.figures import histogram_strip
```

```python
# --- Configuration ---
excel_path = "GENE_data.xlsx"   # path to pipeline output Excel file
gene = "GENE"                   # gene name for plot title
output_path = f"{gene}_histogram_stripplot.png"  # change to .html for interactive output, or .svg
```

```python
# --- Load data ---
scores_df = pd.read_excel(excel_path, sheet_name="scores")
thresh_df = pd.read_excel(excel_path, sheet_name="thresholds")

thresholds = [
    thresh_df["non_functional_threshold"].iloc[0],
    thresh_df["functional_threshold"].iloc[0],
]
print(f"Loaded {len(scores_df)} variants. Thresholds: {thresholds}")
```

```python
# --- Generate plot ---
histogram, stripplot = histogram_strip.make_figures(scores_df, thresholds, gene=gene)
chart = histogram_strip.combine(histogram, stripplot)
chart
```

```python
# --- Save ---
chart.save(output_path)
print(f"Saved: {output_path}")
```
