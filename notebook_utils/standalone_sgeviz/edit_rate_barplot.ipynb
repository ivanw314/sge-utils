# Edit Rate Barplot

Generates a barplot of library edit rates per target and replicate. Each bar represents one replicate's edit rate for a target amplicon, faceted by target. Useful for comparing library efficiency across amplicons.

**Required data:** `edit_rates` sheet from the pipeline output Excel file (`*_data.xlsx`). This sheet is only present if the pipeline was run with an `*editrates*` input file.

The `edit_rates` sheet requires two columns:
- `target_rep`: string in the format `{GENE}_{target}_{rep_string}_{day}` (e.g. `BARD1_X10A_R1R2R3_D05`)
- `edit_rate`: float (0–1)

**Required packages:** `sge-utils` (SimpleSGEViz) installed in the active environment.

**Saving:** PNG/SVG require `vl-convert-python` (`pip install vl-convert-python`). Change the extension to `.html` for interactive output (no extra package needed).

```python
import pandas as pd
from sgeviz.figures import edit_rate_barplot
```

```python
# --- Configuration ---
excel_path = "GENE_data.xlsx"         # path to pipeline output Excel file
gene = "GENE"                         # gene name for plot title
output_path = f"{gene}_edit_rate_barplot.png"  # change to .html for interactive output, or .svg
```

```python
# --- Load data ---
xl = pd.ExcelFile(excel_path)
if "edit_rates" not in xl.sheet_names:
    raise ValueError(
        "No 'edit_rates' sheet found in the Excel file. "
        "Re-run the pipeline with an *editrates* input file to generate this data."
    )

edit_rates_df = pd.read_excel(excel_path, sheet_name="edit_rates")
print(f"Loaded {len(edit_rates_df)} edit rate entries")
```

```python
# --- Generate plot ---
chart = edit_rate_barplot.make_plot(edit_rates_df, gene=gene)
chart
```

```python
# --- Save ---
chart.save(output_path)
print(f"Saved: {output_path}")
```
