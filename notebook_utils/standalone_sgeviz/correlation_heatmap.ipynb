# Replicate Correlation Heatmap

Generates Pearson correlation heatmaps across replicate pairs for SNVs, 3 bp deletions, and combined variants. Useful for assessing replicate reproducibility.

**Required data:** `counts` sheet from the pipeline output Excel file (`*_data.xlsx`).

**Required packages:** `sge-utils` (SimpleSGEViz) installed in the active environment.

**Saving:** PNG/SVG require `vl-convert-python` (`pip install vl-convert-python`). Change the extension to `.html` for interactive output (no extra package needed).

```python
import pandas as pd
from sgeviz.figures import correlation
```

```python
# --- Configuration ---
excel_path = "GENE_data.xlsx"   # path to pipeline output Excel file
gene = "GENE"                   # gene name for plot title
output_path = f"{gene}_correlation_heatmap.png"  # change to .html for interactive output, or .svg
```

```python
# --- Load data ---
counts_df = pd.read_excel(excel_path, sheet_name="counts")
print(f"Loaded counts: {counts_df.shape[0]} variants, {counts_df.shape[1]} columns")
```

```python
# --- Generate plot ---
chart = correlation.make_combined_heatmap(counts_df, gene=gene)
chart
```

```python
# --- Save ---
chart.save(output_path)
print(f"Saved: {output_path}")
```
